## Filtering Lines on The Dataset

```python
imunization_in_brazil = dataset[dataset["country"] == "Brazil"] # filtering only the lines wher country is Brazil
#imunization_in_brazil
imunization_in_brazil[["country", "total_vaccinations", "date", "source_name"]] # Catching less columns
```

```python
# selection lines wher the data are between 2021-02-25 - 2021-02-26
thurday = "2021-02-25"
friday = "2021-02-26"
imunization_in_brazil[(imunization_in_brazil["date"] == thurday)|( imunization_in_brazil["date"] == friday)]
```

## Filtering through labels

```python
# Selection the vaccinated population on Brazil and Argentina in the last friday(2021-02-26)
imunization_br_arg= dataset[(dataset["country"] == "Brazil")|(dataset["country"] == "Argentina")]
imunization_br_arg
```
